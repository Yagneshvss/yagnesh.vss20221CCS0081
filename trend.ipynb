```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01"]),
    "value":[10,15,22,28]
    }
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("Date")
plt.ylabel("values")
plt.title("Time series with thread")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01"]),
    "value":[10,20,17,24,16,18,15,20,17]
    }
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("Date")
plt.ylabel("values")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```
